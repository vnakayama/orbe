```python
qtd=7 #poe os objetos que vc tem além da estrela. 
#No caso, eu tenho 3 objetos, por exemplo, 3 planetas além da estrela.

#quando rodar essa celula, você vai ter um arquivo .txt para cada um dos seus objetos.
#nesses arquivos txt, a primeira coluna é o tempo, depois vem a,e,i,w,nodo
#na celula seguinte a esta, eu abro o arquivo do planeta que eu quero e ploto os dados.


####Rode essa celula sem editar o que está aqui embaixo ################
#para rodar a celula é só clicar em algum lugar aqui do texto e fazer "shift+enter"
file=open('trappist_out.dat','r')
data=file.readlines()
file.close()

def separa_objetos(data, num):
    t,a,e,inc,w,nodo=[],[],[],[],[],[]
    for i in range(0, len(data)):
        obj_tempo=float(data[i].split()[0])
        obj_numero=float(data[i].split()[1])
        obj_a=float(data[i].split()[2])
        obj_e=float(data[i].split()[3])
        obj_i=float(data[i].split()[4])
        obj_w=float(data[i].split()[5])
        obj_nodo=float(data[i].split()[6])
        if obj_numero==num:
            t.append(obj_tempo)
            a.append(obj_a)
            e.append(obj_e)
            inc.append(obj_i)
            w.append(obj_w)
            nodo.append(obj_nodo)
    return [t, a, e, inc, w, nodo]

num=[]
for i in range(0,qtd):
    num.append(i+1)

obj=[]
for i in num:
    obj.append(separa_objetos(data,i)) 

for i in num:
    lista=obj[i-1]
    t=lista[0]
    a=lista[1]
    e=lista[2]
    inc=lista[3]
    w=lista[4]
    nodo=lista[5]
    datafile = open("trappist/data/trappist_"+str(i)+".txt", 'w')
    for j in range(0,len(t)):
        datafile.write(str(t[j])+" "+str(a[j]) + " " + str(e[j]) + " " + str(inc[j]) + " " +str(w[j])+" "+str(nodo[j])+"\n")
    datafile.close()
```

```python
#agora rode esta celula
import matplotlib.pyplot as plt

#plotando dados para o objeto 1
file=open('trappist/data/trappist_7.txt','r')               #se voce quer ver os dados do planeta 2, coloque _2 ao inves de _1
data=file.readlines()
file.close()
t,a,e,i,w,nodo=[],[],[],[],[],[]
for ii in range(0,len(data)):
    t.append(float(data[ii].split()[0]))
    a.append(float(data[ii].split()[1]))
    e.append(float(data[ii].split()[2]))
    i.append(float(data[ii].split()[3]))
    w.append(float(data[ii].split()[4]))
    nodo.append(float(data[ii].split()[5]))


fig, (ax1, ax2, ax3, ax4, ax5) = plt.subplots(5,1,sharex=True)
fig.set_size_inches(10, 17.5)
# fig.suptitle('Mercúrio')

ax1.plot(t,a, color='purple', marker='o', linestyle='')
ax1.set_ylabel('Semi-eixo (UA)')

ax2.plot(t,e, color='purple', marker='.', linestyle='')
ax2.set_ylabel('Excentricidade')

ax3.plot(t,i, color='purple', marker='.', linestyle='')
ax3.set_ylabel('Inclinação (graus)')

ax4.plot(t,w, color='purple', marker='.', linestyle='')
ax4.set_ylabel('w (graus)')

ax5.plot(t,nodo, color='purple', marker='.', linestyle='')
ax5.set_ylabel('Nodo (graus)')
ax5.set_xlabel('Tempo (anos)')

plt.show()
fig.savefig('trappist/img/trappist1-h.png')
```
